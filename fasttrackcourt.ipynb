```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("fasttrackcourt1.csv")
data
ypoints = np.array(data["2020"])
ypoints1 = np.array(data["2021"])
ypoints2 =np.array(data["2022"])
xpoints = np.array(data["State/UT"])
```

```python
data.describe
```

```python
plt.bar(data["State/UT"],data["2020"], color ='green',width = 0.25)
plt.xlabel("State/UT")
plt.ylabel("No. of Functional Court")
plt.title("State-wise No. of Functional Fast Track Court 2020")
```

```python
plt.bar(data["State/UT"],data["2021"], color ='purple',width = 0.25)
plt.xlabel("State/UT")
plt.ylabel("No. of Functional Court")
plt.title("State-wise No. of Functional Fast Track Court 2021")
```

```python
plt.bar(data["State/UT"],data["2022"], color ='red',width = 0.25)
plt.xlabel("State/UT")
plt.ylabel("No. of Functional Court")
plt.title("state-wise No. of Functional Fast Track Court 2022")
```
